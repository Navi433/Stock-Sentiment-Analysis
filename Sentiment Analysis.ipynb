```python
headlines=[]
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
```

```python
ct=CountVectorizer(ngram_range=(2, 2))
```

```python
traindataset=ct.fit_transform(headlines)
```

```python
traindataset
```

```python
classifier.fit(traindataset,train['Label'])
```

```python
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
    testdataset=ct.transform(test_transform)
    predictions=classifier.predict(testdataset)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(predictions,test['Label'])
```
